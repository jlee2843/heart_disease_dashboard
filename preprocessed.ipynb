```python
import pandas as pd
import numpy as np
from ucimlrepo import fetch_ucirepo
```

```python
heart_disease = fetch_ucirepo(id=45)
X = pd.DataFrame(heart_disease["data"]["features"])
y = pd.DataFrame(heart_disease["data"]["targets"])
print(X.shape, y.shape)
```

```python
df = pd.concat([X, y], axis=1)
df.head()
```

```python
df.to_csv("data/raw/heart_disease.csv", index=False)
```

```python
for ind, row in df.iterrows():
    if row["num"] != 0:
        df.loc[ind, "num"] = 1
df.num.value_counts()
```

```python
df["floor_age"] = 0
for ind, row in df.iterrows():
    floor_val = np.floor(row["age"] / 10) * 10
    df.loc[ind, "floor_age"] = floor_val
df.floor_age.value_counts()
```

```python
sex_dict = {1:"male", 0:"female"}
df["sex"] = df["sex"].map(sex_dict)
df
```

```python
cp_dict = {
    1:"typical angina",
    2:"atypical angina",
    3:"non-anginal pain",
    4:"asymptomatic"
}
df["cp"] = df["cp"].map(cp_dict)
```

```python
num_dict = {
    0:"absence",
    1:"presence"
}
df["num"] = df["num"].map(num_dict)
```

```python
fbs_dict = {
    1:"yes",
    0:"no"
}
df["fbs"] = df["fbs"].map(fbs_dict)
```

```python
df.to_csv("data/processed_heart_disease.csv", index=False)
```
